```python
%matplotlib inline
import os
from IPython.display import Image
```

```python
from imageai.Detection import ObjectDetection
```

```python
path = 'C:\Python\\resnet50_coco_best_v2.0.1.h5'
image = 'C:\Python\\image.jpg'
result = 'C:\Python\\result.jpg'
```

```python
detector = ObjectDetection()
```

```python
detector.setModelTypeAsRetinaNet()
detector.setModelPath(path)
detector.loadModel()
```

```python
Image(image)
```

```python
list = detector.detectCustomObjectsFromImage(input_image=image,output_image_path=result,minimum_percentage_probability=20,display_object_name=True,display_percentage_probability=False)
```
