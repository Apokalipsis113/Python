```python
pop = pd.read_csv('data\state-population.csv')
areas = pd.read_csv('data\state-areas.csv')
abbvers = pd.read_csv('data\state-abbrevs.csv')

print(pop.head()) #.head() shows first 5 by default
print(areas.head())
print(abbvers.head())
```

```python
merged = pd.merge(pop,abbvers,how='outer', left_on = 'state/region', right_on='abbreviation')
merged.drop('abbreviation',1)
```

```python
merged.isnull().any()
```

```python
merged[merged['population'].isnull()].head(10)
```

```python
merged.loc[merged['state'].isnull(),'state/region'].unique()
```

```python
merged.loc[merged['state/region'] == 'PR', 'state'] = 'Puerto Rico'
merged.loc[merged['state/region'] == 'USA', 'state'] = 'United States'
merged.isnull().any()
```

```python
final = pd.merge(merged, areas, on = 'state', how = 'left')
final.head(10)
```

```python
final.isnull().any()
```

```python
final['state'][final['area (sq. mi)'].isnull()].unique()
```

```python
final.dropna(inplace=True)
```

```python
final.head(10)
```

```python
final = final.drop('abbreviation',axis=1)
final.head(10)
```

```python
final.isnull().any()
```

```python
data2010 = final.query("year == 2010 & ages == 'total'")
data2010.head(10)
```

```python
data2010.set_index('state',inplace=True)
```

```python
density = data2010['population'] / data2010['area (sq. mi)']
density.sort_values(ascending=False, inplace=True)
#density.head(10)
print(density)
```
