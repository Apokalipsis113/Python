```python
import numpy as np
from PIL import Image
from rawkit.raw import Raw
import os
import rawpy
import PIL
```

```python
directory = 'C:/1/3/'

files = os.listdir(directory)
```

```python
for image in files:
    raw_image = rawpy.imread(directory+image)
    print(directory+image)
    rgb = raw_image.postprocess()
    image = image.replace('.CR2','.jpg')
    print(directory+image)
    PIL.Image.fromarray(rgb).save(directory+image, quality=100, optimize=True)
    raw_image.close()
```
