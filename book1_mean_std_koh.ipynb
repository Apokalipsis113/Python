```python
import numpy as np
import pandas as pd
import xlrd
```

```python
file_name = 'book1.xlsx'
```

```python
data = pd.read_excel(file_name)
source = pd.DataFrame(data)
source
```

```python
for lab  in data["lab"].unique():
    tmp = pd.DataFrame(data.loc[data.lab == lab ])
    for tag in tmp["tag"].unique():
        t = pd.DataFrame(tmp.loc[data.tag == tag])
#         print("lab> ", lab ," tag> ", tag)
#         print(t)
        new_data = pd.DataFrame(columns=tmp.columns)
        new_data = new_data.append({"lab":lab,"tag":tag+"mean","value":t.value.mean()},ignore_index=True)
        new_data = new_data.append({"lab":lab,"tag":tag+"std","value":t.value.std()},ignore_index=True)
        new_data = new_data.append({"lab":lab,"tag":tag+"kohr","value":t.value.std()*t.value.std()},ignore_index=True)
        data = pd.concat([data, new_data],ignore_index=True)
```
