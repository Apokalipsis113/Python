```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
x = np.arange(-10,10,0.001)
```

```python
def Y(x):    
    return 3*(x**2) + 1
def yyy(x,f):
    xex = 0.0000001
    return (f(x+xex)-f(x))/xex
```

```python
res = np.zeros(shape=x.shape[0],dtype=np.float64)
der = np.zeros(shape=x.shape[0],dtype=np.float64)
```

```python
for i in np.arange(x.shape[0]):
    res[i]=Y(x[i])
    der[i]=yyy(x[i],Y)
print(res)
print(der)
```

```python
plt.plot(x,der,'k-')
plt.plot(x,res,'g-');
```
