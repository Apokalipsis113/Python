```python
from pandas_datareader import data
goog = data.DataReader('GOOG',start='2004',end='2019', data_source='stooq')
goog.head()
```

```python
goog = goog['Close']
```

```python
import matplotlib.pyplot as plt
import seaborn; seaborn.set()
%matplotlib inline
```

```python
goog.plot()
```

```python
goog.plot(alpha=0.5, style='-')
goog.resample('BA').mean().plot(style=':')
```

```python

ax = plt.subplots()
data = goog.iloc[:10]
data.resample('B').mean().plot(style='--^')
ax[1].legend(["back-fill", "forward-fill"]);
```

```python
rolling = goog.rolling(365, center=True)
data = pd.DataFrame({'input':goog, 'one-year rolling_mean': rolling.mean(), 'one-year rolling_std': rolling.std()})
ax = data.plot(style=['-','--',':'])
ax.lines[0].set_alpha(0.3)
```

```python
data = pd.read_csv('data/FremomtBridge.csv', index_col='Date', parse_dates=True)
data.head()
```

```python
data.shape
```

```python
data.columns = ['West', 'East']
data['Total'] = data.eval('East + West')
```

```python
data.dropna().describescribe()
```

```python
%matplotlib inline
import seaborn; seaborn.set()
```

```python
data.plot()
plt.ylabel('Houry Bicycle Count');
```

```python
weekly = data.resample('W').sum()
```

```python
weekly.plot(style=[':','--','-'])
plt.ylabel('Weekly Bicycle Count');
```

```python
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':','--','-'])
plt.ylabel('mean hourly count');
```

```python
daily.rolling(50,center=True,win_type='gaussian').sum(std=10).plot(style=[':','--','-']);
```

```python
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks=hourly_ticks, style=[':','-','--']);
```

```python
by_weekly = data.groupby(data.index.dayofweek).mean()
by_weekly.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
by_weekly.plot(style=[':','--','-']);
```

```python
weekend = np.where(data.index.weekday < 5, 'Weekday', 'Weekend') #by day of the week
by_time = data.groupby([weekend, data.index.time]).mean() #by time
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 2, figsize=(14, 5))
by_time.loc['Weekday'].plot(ax=ax[0], title='Weekdays', xticks=hourly_ticks, style=[':', '--', '-'])#.loc instead .ix
by_time.loc['Weekend'].plot(ax=ax[1], title='Weekends', xticks=hourly_ticks, style=[':', '--', '-']);
```
