```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
```

```python
data=np.random.randn(1000)
plt.hist(data);
```

```python
plt.hist(data, bins=30,density='normed',alpha=0.5,histtype='stepfilled', color='steelblue', edgecolor='none');
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled',alpha=0.3,density='normed',bins=100)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

```python
counts, bin_edges = np.histogram(data,bins=10)
print(counts)
```

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x,y = np.random.multivariate_normal(mean, cov, 10000).T
```

```python
plt.hist2d(x,y, bins=30, cmap='Blues')
cd = plt.colorbar()
cd.set_label('counts in bin')
```

```python
counts, xedges, yedges = np.histogram2d(x,y,bins=30)
#print(counts)
```

```python
plt.hexbin(x,y, bins=30, cmap='Blues')
cd = plt.colorbar(label='count in bin')
```

```python
from scipy.stats import gaussian_kde

data = np.vstack([x,y])
kde = gaussian_kde(data)

xgrid = np.linspace(-3.5,3.5,40)
ygrid = np.linspace(-6,6,40)
Xgrid, Ygrid = np.meshgrid(xgrid,ygrid)
Z = kde.evaluate(np.vstack([Xgrid.ravel(),Ygrid.ravel()]))

plt.imshow(Z.reshape(Xgrid.shape), origin='lower',aspect='auto',extent=[-3.5,3.5,-6,6],cmap='Blues')
cd = plt.colorbar()
cd.set_label("density")
```
