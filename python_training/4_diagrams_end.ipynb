```python
%matplotlib notebook
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
plt.plot(x,y,'o', color='k');
```

```python
rng = np.random.RandomState(0)
for marker in ['o','.',',','x','+','v','^','<','>','s','d']:
    plt.plot(rng.rand(5), rng.rand(5),marker, label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0,1.8);
```

```python
plt.plot(x,y,'--ok');
```

```python
plt.plot(x,y,'-p',color='gray',markersize='15',linewidth=4,markerfacecolor='white',markeredgecolor='red',markeredgewidth=2)
plt.ylim(-1.2,1.2);
```

```python
plt.scatter(x,y,marker='o');
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
size = 1000 * rng.rand(100)
plt.scatter(x,y,c=colors,s=size,alpha=0.3,cmap='viridis')
plt.colorbar();
```

```python
from sklearn.datasets import load_iris
iris = load_iris()

features = iris.data.T
plt.scatter(features[0],features[1],alpha=0.2,s=100*features[3],c=iris.target,cmap='viridis')
plt.xlabel=(iris.feature_names[0])
plt.xlabel=(iris.feature_names[1]);
```
