```python
import os
from IPython.display import Image
import pandas as pd
import numpy as np
from imageai.Detection import ObjectDetection
from imageai.Detection import cv2
```

```python
path = 'resnet50_coco_best_v2.0.1.h5'
detector = ObjectDetection()
detector.setModelTypeAsRetinaNet()
detector.setModelPath(path)
detector.loadModel()
```

```python
filename = 'C:/Python/5.jpg'
resultfile = 'C:/Python/result.png'
```

```python
res = pd.DataFrame(detector.detectCustomObjectsFromImage(input_image=filename,
                                                         output_image_path=resultfile, 
                                                         minimum_percentage_probability=5, 
                                                         display_object_name=True, 
                                                         display_percentage_probability=True))
```
